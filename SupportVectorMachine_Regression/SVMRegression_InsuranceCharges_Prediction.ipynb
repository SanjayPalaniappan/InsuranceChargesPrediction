```python
# Importing required python libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Loading the dataset
insurance_data=pd.read_csv('insurance_pre.csv')
insurance_data
```

```python
#Getting the column names of insurance dataset
insurance_data.columns
```

```python
#Converting categorical data into numerical data
ins_data_numerical=pd.get_dummies(insurance_data,drop_first=True)
ins_data_numerical
```

```python
#Getting the column names of insurance dataset, after converting categorical data to numerical 
ins_data_numerical.columns
```

```python
#Storing input and output into separate variables
indep_ins_Data=ins_data_numerical[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
dep_ins_data=ins_data_numerical[['charges']]
```

```python
#Splitting data into train and test
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(indep_ins_Data,dep_ins_data,test_size=1/3,random_state=0)
```

```python
#Applying standardization to bring data into same scale
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
# Building SVM model using train data
from sklearn.svm import SVR
svm_regressor=SVR(kernel='rbf',C=10000)
svm_regressor=svm_regressor.fit(X_train,y_train)
```

```python
#Getting count of support vectors
svm_regressor.n_support_
```

```python
#Getting position of support vectors
svm_regressor.support_
```

```python
#Getting values of support vectors
svm_regressor.support_vectors_
```

```python
#Getting the origin/starting point
svm_regressor.intercept_
```

```python
#Predicting the built model on test data
y_pred=svm_regressor.predict(X_test)
y_pred
```

```python
#Calculating R^2 value
from sklearn.metrics import r2_score
svm_rsquared_value=r2_score(y_test,y_pred)
svm_rsquared_value
```
